```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import sqrt
%matplotlib inline

#
# f = demand
# p = price
# m = marketing
# r = quality

ep = -1.5
em = 1.2
er = 1.5

def f(p, m, r):
    return 10*(p**ep) * (m**em) * (r**er)

p = 1
m = 1
r = 2

x = np.linspace(1,15)
p = x
y = f(p, m ,r)
y2 = x*f(p, m ,r)

fig = plt.figure()
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(14, 10))
fig.tight_layout()
axes[0].plot(x, y, 'r')
axes[1].plot(x, y2, 'r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import sqrt
%matplotlib inline

p = 1
m = 1
r = 1

# production
pd=0.2

x = np.linspace(1,15)
p = x
y = x*np.minimum(f(p, m ,r), pd)

fig = plt.figure()
plt.plot(x,y, 'r')
```
