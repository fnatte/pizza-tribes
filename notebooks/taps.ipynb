```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import sqrt, round
%matplotlib inline

bonus = 1
streak = 0
level = np.arange(1,11,1)

factorPizzas = 70 + 12 * (level ** 1.35) * sqrt(streak+1) * bonus
factorCoins = 35 + 4 * (level ** 1.75) * sqrt(streak+1) * bonus
incrPizzas = round(factorPizzas/5) * 5
incrCoins = round(factorCoins/5) * 5

factor_old = sqrt(level * (streak+1)) * bonus
incrPizzasOld = round(80*factor_old/5) * 5
incrCoinsOld = round(35*factor_old/5) * 5

fig = plt.figure()
plt.plot(level, incrPizzas, 'r')
plt.plot(level, incrCoins, 'b')
plt.plot(level, incrPizzasOld, 'ro')
plt.plot(level, incrCoinsOld, 'bo')

incrCoins[2],incrCoinsOld[2]
# incrPizzas[2], incrPizzasOld[2]
```
