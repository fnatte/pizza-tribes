```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import sqrt, round
%matplotlib inline

def f(x):
    a = 3100
    b = 700
    c = -15
    v = a*(1+x) + b*(1+x)**2 + c*(1+x)**3
    return round(v  / 100) * 100

def f2(x):
    a = 2000
    b = 1500
    c = -12
    v = a*(1+x) + b*(1+x)**2 + c*(1+x)**3
    return round(v  / 100) * 100

x = np.linspace(0,30)
y = f(x)
y2 = f2(x)

fig = plt.figure()
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(14, 10))
fig.tight_layout()
axes[0].plot(x, y, 'r')
axes[1].plot(x, y2, 'r')
plt.show()

print(y)
print(y2)
```
